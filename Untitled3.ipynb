```python
#How many people have heart disease and how many does'nt have heart disease?
df.target.value_counts()
```

```python
df.target.value_counts().plot(kind='bar',color =['blue','green'])
plt.title('Heart Disease Values')
plt.xlabel("1 = Heart Disease, 0 = No Heart Disease")
plt.ylabel('Amount')
```

```python
#2.People of which sex has most heart disease?

df.sex.value_counts()
```

```python
pd.crosstab(df.target, df.sex)
```

```python
sns.countplot(x='target',data=df,hue='sex')
plt.xlabel("0 = No Heart Disease , 1 = Heart Disease")
print("From the sample size we can conclude that more 45% male and more than 75% female have heart disease")
```

```python
#3.What is the most common type of chest pain?

df.cp.value_counts()
```

```python
df.cp.value_counts().plot(kind='pie',figsize=[8,16])
plt.title=('Chest pain')
```

```python
#Most people have type 0 chest pain.
```

```python
#4.People of which sex has which type of chest pain more?

pd.crosstab(df.cp,df.sex)
```

```python
pd.crosstab(df.sex,df.cp).plot(kind='bar',color=['lightskyblue','coral','plum','khaki'])
plt.xlabel("0 = Female , 1 = Male")
```

```python
#Most of male has type 0 while in case of female type 0 and type 2 is almost same
```

```python
#5.People with which chest pain are most pron to have heart disease ?

pd.crosstab(df.cp,df.target)
```

```python
sns.countplot(x='cp',data=df,hue='target')
```

```python
#people wich chest pain of type 0 has the least chance getting heart disease
```

```python
#6.What is the cholestrol level among the people ?

sns.displot(x='chol',data=df,bins=30,kde=True,color='blue');
```

```python
#from this plot we get a clear overview of the cholestrol level among the people in the subset
```

```python
#7.Find any relation between sex and fasting blood sugar level ?

#0 means fbs is above than 120 mg/dl
#1 means fbs is lower than 120 mg/dl
pd.crosstab(df.sex,df.fbs)
```

```python
sns.countplot(x='fbs',data=df,hue='sex')
plt.xlabel("0 = fbs above 120 mg/dl , 1 = fbs lower than 120 mg/dl")
```

```python
#Most people have fbs above 120 mg/dl
#86.53% female have their fbs above 120 mg/dl
#84.43% male have their fbs above 120 mg/dl
```
